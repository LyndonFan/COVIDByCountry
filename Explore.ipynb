```python
import folium
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from folium import plugins
```

```python
df_raw = pd.read_csv("time_series_covid19_confirmed_global.csv")
df_raw.head()
```

```python
df = df_raw[(df_raw["Lat"]!=0) & (df_raw["Long"]!=0)]
```

```python
m = folium.Map(location=[0, 0],tiles='Stamen Toner', zoom_start=1.5)
for index, row in df.iterrows():
    name = (row["Province/State"] + ", ") if type(row["Province/State"])==str else ""
    name += row["Country/Region"]
    folium.Circle(
        location = [row["Lat"],row["Long"]],
        radius = np.power(row["4/15/20"],0.5)*1000,
        popup = name,
        stroke = False,
        fill = True,
        fill_color = "dark_red"
    ).add_to(m)
m
```

```python
df_cul = df.copy()
for i in range(5,len(df.columns)):
    df_cul.iloc[:,i] += df_cul.iloc[:,i-1]
df_cul.sort_values(by="4/15/20",axis=0,ascending = False,inplace=True)
```

```python
m_cul = folium.Map(location=[0, 0],tiles='Stamen Terrain', zoom_start=1.5)
for index, row in df_cul.iterrows():
    name = (row["Province/State"] + ", ") if type(row["Province/State"])==str else ""
    name += row["Country/Region"]
    folium.Circle(
        location = [row["Lat"],row["Long"]],
        radius = np.power(row["4/15/20"],0.5)*500,
        popup = "{}: {} cases".format(name,row["4/15/20"]),
        stroke = False,
        fill = True,
        color = "red",
        fill_opacity = 0.4
    ).add_to(m_cul)
m_cul
```
